```python
import pandas as pd
df=pd.read_csv("hindi_songs.csv")
```

```python
df.head()
```

```python
df.dropna()
```

```python
df.drop_duplicates()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['id','album_name','is_hindi'],inplace=True)
```

```python
artist_counts = df['artists'].value_counts()
```

```python
artist_counts
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,7))
artist_counts.head(30).plot(kind='bar')
plt.title('Top 20 Artists by Number of Songs')
plt.xlabel('Artist Name')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45)
plt.show()
```

```python
print(df['duration_ms'].head())
```

```python
min_duration = df['duration_ms'].min()
max_duration = df['duration_ms'].max()
duration_range = max_duration - min_duration

print(f"The range of song durations is from {min_duration} seconds to {max_duration} seconds, with a range of {duration_range} seconds.")
```

```python
df['duration_seconds'] = df['duration_ms'] / 1000
```

```python
min_duration = df['duration_seconds'].min()
max_duration = df['duration_seconds'].max()
duration_range = max_duration - min_duration

print(f"The range of song durations is from {min_duration} to {max_duration} seconds, with a range of {duration_range} seconds.")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.hist(df['duration_seconds'], bins=15, color='skyblue', edgecolor='black')
plt.title('Distribution of Song Durations (in seconds)')
plt.xlabel('Duration (seconds)')
plt.ylabel('Number of Songs')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
df.head()
```

```python
features = ['danceability', 'energy', 'acousticness', 'liveness', 'valence']

plt.figure(figsize=(14, 8))
for i, feature in enumerate(features, 1):
    plt.subplot(2, 3, i)
    plt.hist(df[feature], bins=20, color='orange', edgecolor='black')
    plt.title(f'Distribution of {feature.capitalize()}')
    plt.xlabel(f'{feature.capitalize()}')
    plt.ylabel('Number of Songs')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
plt.scatter(df['danceability'], df['energy'], color='green')
plt.title('Danceability vs. Energy')
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(df['danceability'], df['valence'], color='purple')
plt.title('Danceability vs. Valence')
plt.xlabel('Danceability')
plt.ylabel('Valence')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

top_20_artists = df['artists'].value_counts().nlargest(20).index
df_top_20 = df[df['artists'].isin(top_20_artists)]
plt.figure(figsize=(12, 6))
sns.boxplot(x='artists', y='danceability', data=df_top_20, order=top_20_artists)
plt.title('Danceability Distribution by Top 20 Artists')
plt.xlabel('Artist')
plt.ylabel('Danceability')
plt.xticks(rotation=90)
plt.show()
```

```python
music_features = ['danceability', 'energy', 'loudness', 'speechiness', 
                  'acousticness', 'instrumentalness', 'liveness', 'valence', 
                  'tempo']
corr_matrix = df[music_features].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Music Features')
plt.show()
```
